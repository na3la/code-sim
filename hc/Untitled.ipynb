```python
import matplotlib.pyplot as plt
import scipy.spatial.distance as chec 
import pandas as pd
%matplotlib inline
import numpy as np
```

```python
srcdat = pd.read_csv('/home/anon/notebooks/code-sim/tokenizer/src/oldcsvs/testfeatures.csv')
```

```python
srcdat.shape
```

```python
srcdat.head()
```

```python
dat = srcdat.iloc[:,1:].values
```

```python
import scipy.cluster.hierarchy as shc
dat
```

```python
plt.figure(figsize=(10,7))
plt.title("Src Dendrograms")
dend = shc.dendrogram(shc.linkage(chec.pdist(dat), method = 'complete'))
```

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
cluster = AgglomerativeClustering(n_clusters=4, affinity='euclidean', linkage='ward')
cluster.fit_predict(dat)
```

```python
plt.figure(figsize=(10,7))
plt.title('Javahw1')
plt.xlabel('num_unique_match_lengths')
plt.ylabel('num_match_char')
plt.scatter(dat[:,1], dat[:,0], c=cluster.labels_, cmap='rainbow');
```

```python
plt.figure(figsize=(10,7))
plt.title('Javahw1')
plt.xlabel('num_unique_match_lengths')
plt.ylabel('num_match_char')
plt.scatter(dat[:,1], dat[:,0], c=cluster.labels_, cmap='rainbow');
for label, y, x in zip(srcdat['fileID'], srcdat['maxmatch'], srcdat['num_vals']):
    plt.annotate(
        label, xy =(x,y),
        xytext=(-3,3),
        textcoords='offset points',
        ha='right',
        va='bottom')
plt.show();
```
