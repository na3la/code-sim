```python
## Source Code print

# Hash to decrease num comparisons? hash tokenized version?
def prntcodenhash(indx, hwn):
    
    if hwn == 1:
        filename = srcdat.columns[indx]
        f = '/home/anon/notebooks/code-sim/tokenizer/javahw1/fixed/f' + filename + \
        '_assignsubmission_file_InstTable.java'
    elif hwn == 2.1:
        filename = srcdathw2eq.columns[indx]
        f = '/home/anon/notebooks/code-sim/tokenizer/javahw2/fixed/eqnumcondition/f'+ filename + \
        '_assignsubmission_file_EqNumCondition.java'
    elif hwn == 2.2:
        filename = srcdathw2or.columns[indx]
        f = '/home/anon/notebooks/code-sim/tokenizer/javahw2/fixed/orcondition/f' + filename + \
        '_assignsubmission_file_OrCondition.java'
    elif hwn == 2.3:
        filename = srcdathw2select.columns[indx]
        f = '/home/anon/notebooks/code-sim/tokenizer/javahw2/fixed/selectquery/f' + filename + \
        '_assignsubmission_file_SelectQuery.java'
    elif hwn == 3:
        filename = srcdathw3.columns[indx]
        f = '/home/anon/notebooks/code-sim/tokenizer/javahw3/fixed/f' + filename + \
        '_assignsubmission_file_SelectQuery.java'
    
    fi = ""
    with open(f, 'r') as ff:
        for x in ff:
            fi += x
    
    print(filename + " " + ': \n')
    print(hash(''.join(fi.split())))
    print(fi + '\n')
```
