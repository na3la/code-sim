```python
token_ltt_ldd = pd.read_csv("../../experimental/code-sim/token_ltt_ldd.csv")
token_avg_ltt = pd.read_csv("../../experimental/code-sim/token_avg_ltt.csv")
ltt_ldd = pd.read_csv("../../experimental/code-sim/completet_ltt_ldd.csv")
avg_ltt = pd.read_csv("../../experimental/code-sim/completet_avg_ltt.csv")
```

```python
def format(dat):
    dat = dat.rename(columns={'Unnamed: 0':'user_id_row'})
    dat.set_index('user_id_row', inplace=True)
    dat.columns = dat.index
    dat = dat.rename_axis('user_id_col', axis=1)
    return dat
```

```python

avg_ltt = format(avg_ltt)
token_ltt_ldd = format(token_ltt_ldd)
token_avg_ltt = format(token_avg_ltt)
ltt_ldd = format(ltt_ldd)
```

```python
def pairs_list(dist_mat):
    pairs = []
    for x in dist_mat.index:
        t = []
        for y in dist_mat.columns:
            if not x ==y:
                t.append((dist_mat.index.get_loc(x), dist_mat.columns.get_loc(y),))
            else:
                pairs.append(t)
                break
    return pairs

def create_data_frame(dist_mat):
    return pd.DataFrame(index=dist_mat.index, columns=dist_mat.columns)

def insert_data(pairs, old_df, new_df):
    for row in pairs:
        if not len(row):
            continue
        for pair in row:
            new_df.iat[pair[0], pair[1]] = old_df.iloc[pair]
```

```python
def gen_dict(df):
    return df[df == 0].stack().index.tolist()
```

```python
ltt_ldd_pairs = pairs_list(ltt_ldd)
ltt_ldd_tri_mat = create_data_frame(ltt_ldd).astype(float)
insert_data(ltt_ldd_pairs, ltt_ldd, ltt_ldd_tri_mat)
zero_v_d_ltt_ldd = gen_dict(ltt_ldd_tri_mat)

token_ltt_ldd_pairs = pairs_list(token_ltt_ldd)
token_ltt_ldd_tri_mat = create_data_frame(token_ltt_ldd).astype(float)
insert_data(token_ltt_ldd_pairs, token_ltt_ldd, token_ltt_ldd_tri_mat)
zero_v_d_token_ltt_ldd = gen_dict(token_ltt_ldd_tri_mat)


avg_ltt_pairs = pairs_list(avg_ltt)
avg_ltt_tri_mat = create_data_frame(avg_ltt).astype(float)
insert_data(avg_ltt_pairs, avg_ltt, avg_ltt_tri_mat)
zero_v_d_avg_ltt = gen_dict(avg_ltt_tri_mat)

token_avg_ltt_pairs = pairs_list(token_avg_ltt)
token_avg_ltt_tri_mat = create_data_frame(token_avg_ltt).astype(float)
insert_data(token_avg_ltt_pairs, token_avg_ltt, token_avg_ltt_tri_mat)
zero_v_d_token_avg_ltt = gen_dict(token_avg_ltt_tri_mat)
```
