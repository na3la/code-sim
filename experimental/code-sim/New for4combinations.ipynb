```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import scipy.spatial.distance as chec 
import scipy.cluster.hierarchy as shc
```

```python
# read csv files.
# dist_1 --> (avg # tokens/cst score)-1
# dist_2 --> (1 - (2*max_length_match)/(sum(prog1.len + prog2.len)))
# eq_1 --> token check
# eq_2 --> token + value check ****this is equivalent to only value check****

dist_1_eq_1 = pd.read_csv("FULLWORD.csv")
dist_1_eq_2 = pd.read_csv("FULLWORD3.csv")
dist_2_eq_1 = pd.read_csv("FULLWORDFULLEQUALITY.csv")
dist_2_eq_2 = pd.read_csv("FULLWORDFULLEQUALITY3.csv")
```

```python
dat = [dist_1_eq_1, dist_1_eq_2, dist_2_eq_1, dist_2_eq_2]
for x in range(len(dat)):
    dat[x] = dat[x].rename(columns={'Unnamed: 0':'user_id'})
    dat[x] = dat[x].set_index('user_id')
```

```python
dat[0].head()
```

```python
plt.figure(figsize=(10,10))
plt.title("Javahw1 - (Average Program Size/Total Matching Tokens) - 1 and Token Check")
plt.ylabel("User ID")
plt.xlabel("Distance")
dend = shc.dendrogram(shc.linkage(chec.pdist(dat[0]), method='complete'), labels=dat[0].columns, orientation='right')
```

```python
plt.figure(figsize=(10,10))
plt.title("Javahw1 - (Average Program Size/Total Matching Tokens) - 1 and Token + Value Check")
plt.ylabel("User ID")
plt.xlabel("Distance")
dend = shc.dendrogram(shc.linkage(chec.pdist(dat[1]), method='complete'), labels=dat[1].columns, orientation='right')
```

```python
plt.figure(figsize=(10,10))
plt.title("Javahw1 - (1 - (2*max_length_match)/(sum(prog1.len + prog2.len))) and Token Check")
plt.ylabel("User ID")
plt.xlabel("Distance")
dend = shc.dendrogram(shc.linkage(chec.pdist(dat[2]), method='complete'), labels=dat[2].columns, orientation='right')
```

```python
plt.figure(figsize=(10,10))
plt.title("Javahw1 - (1 - (2*max_length_match)/(sum(prog1.len + prog2.len))) and Token + Value Check")
plt.ylabel("User ID")
plt.xlabel("Distance")
dend = shc.dendrogram(shc.linkage(chec.pdist(dat[3]), method='complete'), labels=dat[3].columns, orientation='right')
```
