```python
import pandas as pd
import matplotlib
import numpy as np
```

```python
c_dat = pd.read_csv('C_token1.csv')
c_dat_2 = pd.read_csv('C_token2.csv')
c_diff = pd.read_csv('C_diff_ratio')
```

```python
def format_df(df):
    df.index = df["Unnamed: 0"]
    df.rename_axis(index="uuid_row", columns="uuid_col", inplace=True)
    df.drop(columns="Unnamed: 0", inplace=True)
```

```python
format_df(c_dat)
format_df(c_dat_2)
format_df(c_diff)
```
