```python
c_dat = pd.read_csv('ltt_ldd.csv')
c_dat_2 = pd.read_csv('avg_ltt.csv')
c_diff = pd.read_csv('diff.csv')
```

```python
def format_df(df):
    df.index = df["Unnamed: 0"]
    df.rename_axis(index="uuid_row", columns="uuid_col", inplace=True)
    df.drop(columns="Unnamed: 0", inplace=True)
```

```python
format_df(c_dat)
format_df(c_dat_2)
format_df(c_diff)
```
